```r
#Stacked Violin Plot
#https://divingintogeneticsandgenomics.rbind.io/post/stacked-violin-plot-for-visualizing-single-cell-data-in-seurat/

## remove the x-axis text and tick
## plot.margin to adjust the white space between each plot.
## ... pass any arguments to VlnPlot in Seurat
modify_vlnplot<- function(obj, 
                          feature, 
                          pt.size = 0, 
                          plot.margin = unit(c(-0.75, 0, -0.75, 0), "cm"),
                          ...) {
  p<- VlnPlot(obj, features = feature, pt.size = pt.size, ... )  + 
    xlab("") + ylab(feature) + ggtitle("") + 
    theme(legend.position = "none", 
          plot.title= element_blank(),
          axis.title.x = element_blank(),
          axis.text.x = element_blank(), 
          axis.ticks.x = element_blank(), 
          axis.title.y = element_text(family = "Arial", size = 6, angle = 0, colour = "black", face = "plain"), 
          axis.text.y = element_blank(), 
          plot.margin = plot.margin ) 
  return(p)
}
## extract the max value of the y axis
extract_max<- function(p){
  ymax<- max(ggplot_build(p)$layout$panel_scales_y[[1]]$range$range)
  return(ceiling(ymax))
}
## main function
StackedVlnPlot<- function(obj, features,
                          pt.size = 0, 
                          plot.margin = unit(c(-0.05, 0, -0.05, 0), "cm"),
                          ...) {
  
  plot_list<- purrr::map(features, function(x) modify_vlnplot(obj = obj,feature = x, ...))
  
  # Add back x-axis title to bottom plot. patchwork is going to support this?
  plot_list[[length(plot_list)]]<- plot_list[[length(plot_list)]] +
    theme(axis.text.x=element_text(family = "Arial", size=8, face = "plain"),
          axis.ticks.x = element_line())
  
  
  # change the y-axis tick to only max value 
  ymaxs<- purrr::map_dbl(plot_list, extract_max)
  plot_list<- purrr::map2(plot_list, ymaxs, function(x,y) x + 
                            scale_y_continuous(breaks = c(y)) + 
                            expand_limits(y = y))
  
  p<- patchwork::wrap_plots(plotlist = plot_list, ncol = 1, heights = 1)
  return(p)
}
```

```r
##Calculate clonality of TCR repertoire in each cluster
Clonality <- function(dir.input, file.name){
  name.input <- str_c(dir.input, file.name, sep = "/")
  d <- read.table(name.input, header = TRUE)
  clust.array <- names(dplyr::select(d, contains("X")))
  output <- data.frame()
  for(i in 1:length(clust.array)){
    clust <- clust.array[i]
    d_sub <- select(d, clust)
    names(d_sub) <- "count"
    d_sub <- filter(d_sub, count > 0)
    d_sub$freq <- d_sub$count / sum(d_sub$count)
    
    ##Calculate median and count clonotypes
    l <- length(d_sub$freq)
    count <- sum(d_sub$count)
    
    #Calculate 1-Pielou clonality
    e <- log(d_sub$freq)
    pi <- 1+sum(d_sub$freq * e)/log(length(d_sub$freq))
    
    out <- c(l, count, pi)
    
    output <- rbind(output, out)
  }
  names(output) <- c("clones", "cells", "Pielou")
  output$cluster <- clust.array
  output$ID <- str_split(file.name, pattern = "\\.", simplify = TRUE)[1]
  
  return(output)
}
```

```r
##Calculate cluster bias of clones
ClusterBias <- function(dir.input, file.name, th){
  name.input <- str_c(dir.input, file.name, sep = "/")
  d <- read.table(name.input, header = TRUE)
  #Extract tables for cluster distribution
  d <- dplyr::filter(d, temp_count >= th) %>% dplyr::select(contains("X"))
  count.clust <- 1/ncol(d)
  bias <- apply(d, 1, function(x){return(max(x)/sum(x))})
  
  #Calculate frequency table of cluster bias
  h <- hist(bias, breaks=seq(count.clust, 1, 0.1))
  n <- length(h$counts) #number of classes
  class_names <- NULL 
  for(i in 1:n) {
    class_names[i] <- paste(h$breaks[i], "-", h$breaks[i+1])
  }
  frequency_table <- data.frame(class=class_names, frequency=h$counts)
  frequency_table$proportion <- frequency_table$frequency / sum(frequency_table$frequency)
  frequency_table$name <- str_split(file.name, pattern = "\\.", simplify = TRUE)[1] 
  
  return(frequency_table)
}
```

```r
##Plot clones in dimentional reduction plot (DimPlot)
#seu: Seurat object / tmp2: identifier of plotting cells

PlotClones = function(seu, tmp2, name.output, meta.data, clone_plot, reduction, width, height, size){
    #Exctract names for cells in plot
    tmp_raw <- as.vector(row.names(meta.data))
    #Extract colors for plot
    p2 = Seurat::DimPlot(object = seu, reduction = reduction, pt.size = 0.1)
    orig_ident_build = ggplot2::ggplot_build(p2)
    orig_ident_build = orig_ident_build$data[[1]]
    orig_ident_build =  orig_ident_build[order(orig_ident_build$group), ]
    ident.cols = unique(orig_ident_build$colour) # Get a vector of unique colors
    names(ident.cols)=c(0:(max(as.numeric(seu@active.ident)-1)))

    #Extract x- and y- axis information for DimPlot
    xmin <- min(orig_ident_build$x)-0.5
    xmax <- max(orig_ident_build$x)+0.5
    ymin <- min(orig_ident_build$y)-0.5
    ymax <- max(orig_ident_build$y)+0.5
    x_label_min <- floor(xmin/10)*10
    x_label_max <- ceiling(xmax/10)*10
    y_label_min <- floor(ymin/10)*10
    y_label_max <- ceiling(ymax/10)*10

    #Dimplot
    p1 <- DimPlot(object = seu, label = FALSE,
                    cells = tmp2,
                    reduction = reduction,
                    cols = ident.cols,
                    pt.size = size) + 
        NoLegend() +
        scale_x_continuous(limits = c(xmin,xmax), breaks= seq(x_label_min,x_label_max,10)) +
        scale_y_continuous(limits = c(ymin,ymax), breaks= seq(y_label_min,y_label_max,10)) +
        theme(axis.title = element_blank(), 
              axis.text = element_blank(),
             axis.line = element_line(size = 0.2))
    
      save_plot(file = name.output, plot_grid(p1, ncol=1, nrow=1), device="tiff", 
                units="in", dpi = 300, base_width = width, base_height = height, limitsize=FALSE)
}
```

```r
#Figure 7F: Summarize cluster distribution of WGCNA patterns
#Input layer
dir.input <- 'scTCR_analysis'
dir.output <- "7F"
cores <- 12 
Tcells <- c("CD4")

#################################### Processing layer ####################################################
###Define functions
createEmptyDf = function( nrow, ncol, colnames = c() ){
  data.frame( matrix( vector(), nrow, ncol, dimnames = list( c(), colnames ) ) )
}
tableread_fast = function(i, header=TRUE, quote="", sep="\t"){
  tmp = fread(i, header=header, sep=sep, quote=quote, nThread=32)
  tmp = as.data.frame(tmp)
  return(tmp)
}

##Main functions
WGCNA_SCT_summarize = function(file.name, dir.input){
  #load
  name.input <- str_c(dir.input, file.name, sep = "/")
  d <- tableread_fast(name.input, header = TRUE, quote="\"", sep=",")
  
  out.table <- data.frame()
  for(Tcell in Tcells){
    #Prepare data frame
    d_sub2 <- dplyr::select(d, contains("X")) %>% dplyr::select(-c("class.x"))
    d_sub2$class <- d$class.x
    
    #Summarize
    d_sub2 <- group_by(d_sub2, class)
    table <- summarise_each(d_sub2, funs(sum), starts_with("X"))
    table$Tcell <- Tcell
    
    out.table <- rbind(out.table, table)
  }
  
  out.table$ID <- str_split(file.name, pattern = "\\.", simplify = TRUE)[1] 
  
  return(out.table)
}

###Main modules
dir.create(dir.output)

files  <- list.files(dir.input, pattern="CD4.clone_within_cluster.WGCNA.AIM")
t<-proc.time()
cl <- makeCluster(cores)
registerDoParallel(cl)   
out <- foreach(file.name = files, .combine = "rbind",
               .packages=c("stringr", "data.table", "dplyr")) %dopar% {WGCNA_SCT_summarize(file.name, dir.input)}
stopCluster(cl)
proc.time()-t

name.output <- str_c(dir.output, "7F.WGCNA.Distribution.csv", sep = "/")
write.csv(out, name.output, row.names = F)
```

```r
#Figure 7G: Heatmap of cluster distribution of clones (CD4)

dir.name <- "7G"
dir.input <- "scTCR_analysis" #Directory for clone table of scTCR
IDs <- c("002", "027")
Tcell <- "CD4"
th <- 10 #threshold for clones size (count of cells per clone) to plot

##################### Processing layer #############################################################
createEmptyDf = function( nrow, ncol, colnames = c() ){
  data.frame( matrix( vector(), nrow, ncol, dimnames = list( c(), colnames ) ) )
}

dir.create(dir.name)

d <- data.frame()
#Load data
for(ID in IDs){
  #Call clone table
  name.input <- str_c(dir.input, ID, sep = "/") %>% str_c("CD4.clone_within_cluster.WGCNA.AIM.csv", sep = ".")
  clone_within_cluster <- read.csv(name.input, header = TRUE)
    
  #Extract clones: More than X cells per clone
  clone_plot <- dplyr::filter(clone_within_cluster, temp_count >= th)
  
  #Calculate distribution of clones
  d_clust_count <- dplyr::select(clone_plot, starts_with("X"))
  d_clust_prop <- createEmptyDf(nrow(d_clust_count), ncol(d_clust_count), colnames = names(d_clust_count) ) 
  for(i in 1:ncol(d_clust_count)){
    d_clust_prop[,i] <- mapply(function(x,y){return(x/y)}, d_clust_count[,i], clone_plot$temp_count)
  }
  #Append clone information
  d_clust_prop$wgcna <- clone_plot$class.x
  d_clust_prop$AIM <- clone_plot$class.y
  d_clust_prop$ID <- ID
  
  d <- rbind(d, d_clust_prop)
}

#Only analyze CD8 clusters
d_heatmap <- t(as.matrix(dplyr::select(d, c("X0", "X1", "X2", "X3"))))
row.names(d_heatmap) <- c("CM", "EM", "Act", "Treg")

#Output
ppi <- 600
name.output <- str_c(dir.name, Tcell, sep = "/") %>% str_c("clone.WGCNA.heatmap.tiff", sep = ".")
tiff(name.output, width=3.4*ppi, height=1.9*ppi, res=ppi)

ha <- HeatmapAnnotation(df = data.frame(ID = d$ID, WGCNA = d$wgcna),
                        col = list(ID = c("002"="black", "027"="white"),
                                   WGCNA = c("0"="gray", "singlet"="black",
                                             "green" = "turquoise", "turquoise" = "darkgreen", "brown" = "orange", 
                                             "blue" = "gray", "red" = "gray", "yellow" = "gray", 
                                             "invgreen" = "gray", "invturquoise" = "gray", "invbrown" = "gray",
                                             "invyellow" = "gray", "invblue" = "gray", "invred" = "gray")),
                        annotation_name_gp = gpar(fontsize=8, family="Arial"),
                        simple_anno_size = unit(0.4, "cm"),
                        show_legend = c(FALSE, FALSE))
#                        annotation_legend_param =list(labels_gp = gpar(fontsize=6, family="Arial"),
#                                                      title_gp = gpar(fontsize=8, family="Arial"),
#                                                      legend_direction = "vertical"))

p <- Heatmap(d_heatmap, name = "name",
             top_annotation = ha,
             column_dend_height = unit(0.5, "cm"),
             column_names_gp = gpar(fontsize = 6, family="Arial"),
             show_column_names = FALSE,
             show_row_names = TRUE,
             cluster_rows = FALSE,
             row_names_gp = gpar(fontsize=8, family="Arial"),
             heatmap_legend_param =list(labels_gp = gpar(fontsize=6, family="Arial"),
                                        title = "Prop.", 
                                        title_position = "topcenter",
                                        legend_width = unit(2, "cm"),
                                        title_gp = gpar(fontsize=8, family="Arial"),
                                        legend_direction = "vertical"))
q <- draw(p, annotation_legend_side = "right", heatmap_legend_side = "right")
print(q)

dev.off()
```
